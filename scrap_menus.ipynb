```python
driver = start_firefox('https://www.amazon.fr/gp/bestsellers/?ref_=nav_cs_bestsellers')
```

```python
menus = find_all(S('//html/body/div[1]/div[1]/div[2]/div/div/div/div[2]/div/div[2]/div/div/div[2]/div/a'))
```

```python
cat1 = [cat.web_element.text for cat in menus]
links = [cat.web_element.get_attribute('href') for cat in menus]
```

```python
df_cat1 = pd.DataFrame({'Category': cat1, 'Link': links})
df_cat1
```

```python
list_of_df = []

# boucler sur le df
for category, link in zip(df_cat1['Category'], df_cat1['Link']):
    print(category, link)
    go_to(link)
    
    # création des listes de données scrapées
    sub_categories = find_all(S('//html/body/div[1]/div[2]/div/div/div[2]/div/div/div[2]/div[2]/div/a'))
    cat2_text = [cat.web_element.text for cat in sub_categories]
    cat2_links = [cat.web_element.get_attribute('href') for cat in sub_categories]
    cat1 = [category for i in range(len(cat2_text))]
    
    # création d'un df temporaire par page scrapée
    df_to_append = pd.DataFrame({
        'cat1' : cat1,
        'cat2' : cat2_text,
        'cat2_links' : cat2_links
    })
    # ajout du df_to_append dans la liste des df
    list_of_df.append(df_to_append)
    
    
df_with_categories = pd.concat(list_of_df)
df_with_categories
```

# export du df avec toutes les catégories à scraper

```python
df_with_categories.to_csv('datasets/amazon_categories.csv', index = False, encoding = 'utf-8')
```
